# Création d'un fichier unique avec les informations historiques (2015 à 2017) et théoriques

```python
import pandas as pd
import numpy as np
import os
os.chdir("")
```

## Indication des fichiers qui vont être utilisés

```python
file_origine='line_p.csv'
file_compare_verano='crono_verano_P.csv'
file_compare_invierno='crono_invierno_P.csv'
```

## Lecture des fichier et intégration dans des Dataframes

```python
#Lecture du fichier origine et mise dans un dataframe
df_origine=pd.read_csv(file_origine,delimiter=';', low_memory=False)
#Affichage des 10 premières lignes du dataframe
df_origine.head(10)
#df_origine.shape
```

```python
#Lecture du 1er fichier comparaison et mise dans un dataframe
df_compare_verano=pd.read_csv(file_compare_verano,delimiter=';', low_memory=False)
#Affichage des 10 premières lignes du dataframe
df_compare_verano.head(10)
#df_compare_verano.shape
```

```python
#Lecture du 2ème fichier comparaison et mise dans un dataframe
df_compare_invierno=pd.read_csv(file_compare_verano,delimiter=';', low_memory=False)
#Affichage des 10 premières lignes du dataframe
df_compare_invierno.head(10)
#df_compare_invierno.shape
```

## Concaténation des deux fichiers de comparaison des horaires théoriques

```python
df_compare_invierno['SAISON']='INVIERNO'
df_compare_verano['SAISON']='VERANO'

#concaténation des 2 fichiers de comparaison
df_compare=pd.concat([df_compare_invierno,df_compare_verano],ignore_index=True)

#renommage de la colonne ORDEN pour effectuer le merge
df_compare=df_compare.rename(columns={'ORDEN':'FR1_REGIST'})

df_compare.head(10)
#df_compare.shape
```

## Classement des fichiers selon la date et les numéros d'ordre (REGIST)

```python
#trier le dataframe en fonction de 'FR1_FECHA' et 'FR1_REGIST'
df_origine=df_origine.sort_values(by=['FR1_FECHA','FR1_REGIST'])
```

## Rajout d'une colonne indiquant le type de jour

```python
#Rajout d'une colonne indiquant le type de jour
df_origine['TIPO_DIA']=np.where(df_origine['FR1_TIPO']=='F','DOMINGO',
                               np.where(df_origine['FR1_TIPO']=='D','DOMINGO',
                                        np.where(df_origine['FR1_TIPO']=='S','SABADO',
                                                 np.where(df_origine['FR1_TIPO']=='H','HABIL','AUTRE'))))
#Affichage des 10 premières lignes du dataframe
#Compter les autres
pd.value_counts(df_origine['TIPO_DIA'].values, sort=False)
#df_origine.head(10)
```

## Modification du type des colonnes contenant des dates au format date

```python
from datetime import date, datetime

df_origine['FR1_FECHA'] = pd.to_datetime(df_origine['FR1_FECHA'], format="%d/%m/%Y")
df_origine.dtypes
```

## Rajout d'une colonne indiquant le type de saison (VERANO ou INVIERNO)

```python
from datetime import date, datetime
date1=datetime(2015,1,1,0,0,0)
date2=datetime(2015,3,1,0,0,0)
date3=datetime(2015,12,23,0,0,0)
date4=datetime(2016,3,1,0,0,0)
date5=datetime(2016,12,23,0,0,0)
date6=datetime(2017,3,1,0,0,0)
date7=datetime(2017,12,23,0,0,0)
date8=datetime(2018,3,1,0,0,0)

df_origine['SAISON']=np.where(np.logical_and(df_origine['FR1_FECHA']>=date1,df_origine['FR1_FECHA']<date2),'VERANO',
                              np.where(np.logical_and(df_origine['FR1_FECHA']>=date2,df_origine['FR1_FECHA']<date3),'INVIERNO',
                                       np.where(np.logical_and(df_origine['FR1_FECHA']>=date3,df_origine['FR1_FECHA']<date4),'VERANO',
                                                np.where(np.logical_and(df_origine['FR1_FECHA']>=date4,df_origine['FR1_FECHA']<date5),'INVIERNO',
                                                         np.where(np.logical_and(df_origine['FR1_FECHA']>=date5,df_origine['FR1_FECHA']<date6),'VERANO',
                                                                  np.where(np.logical_and(df_origine['FR1_FECHA']>=date6,df_origine['FR1_FECHA']<date7),'INVIERNO',
                                                                           np.where(np.logical_and(df_origine['FR1_FECHA']>=date7,df_origine['FR1_FECHA']<date8),'VERANO','AUTRE')))))))

#Vérifier inversion mois/jour
df_origine.head(10)
```

## Ajout d'une colonne indiquant l'année en question (en ftc de la date)

```python
df_origine['PERIODO']=df_origine['FR1_FECHA'].dt.year
#df_origine.shape
df_origine.head(10)
```

## Merge des 2 fichiers avec les colonnes SAISON, PERIODO, TIPO_DIA et FR1_REGIST

```python
df_final=pd.merge(df_origine,df_compare,on=['SAISON','PERIODO','TIPO_DIA','FR1_REGIST'],how='left')
df_final=df_final.sort_values(by=['FR1_FECHA','FR1_REGIST'])
df_final.shape
```

## Export des données sous CSV

```python
export_tot='BA_' + df_final.iloc[1,1] + '_donnees_preparees.csv'
print(export_tot)
df_final.to_csv(export_tot,sep=';',index=False)
```
