# Calcul du KPI des interruptions

```python
import pandas as pd
import numpy as np
import os
import datetime
os.chdir("")
```

## Indication du fichier que l'on va utiliser

```python
file_origine='BA_Nb_Depart_2019-2020.xlsx'
```

## Lecture du fichier et intégration dans des Dataframes

```python
#Lecture du fichier origine
xl=pd.ExcelFile(file_origine)
#Mise dans un dataframe
df_origine=xl.parse("Data")
#Affichage des 10 premières lignes du dataframe
df_origine.head(10)
```

## Affichage pour rappel du type de données

```python
df_origine.dtypes
```

## Simulation des Interruptions sur 2 ans en utilisant les lois de probabilité

```python
import scipy.stats as st

#Loi de probabilité de la ligne A
df_origine.insert(5,'A_Int_B100',st.exponnorm.rvs(2271.3520030889576, -0.002586904698236756, 0.0007234200726931889, size=731))

#Loi de probabilité de la ligne B
df_origine.insert(7,'B_Int_B100',st.genexpon.rvs(2.3707480090151876, 1.822706440989477e-08, 1.601427692982724, -5.3413762848371975e-08, 3.6673022564923476, size=731))

#Loi de probabilité de la ligne C
df_origine.insert(9,'C_Int_B100',st.wald.rvs(-0.2116701131651372, 0.7304669036293392, size=731))

#Loi de probabilité de la ligne D
df_origine.insert(11,'D_Int_B100',st.exponnorm.rvs(2920.2368038086333, -0.0019276999851486065, 0.0005550819703773408, size=731))

#Loi de probabilité de la ligne E
df_origine.insert(13,'E_Int_B100',st.mielke.rvs(0.7479261485668828, 1.3493198675745806, -5.594979913033181e-32, 1.074308957921079, size=731))

#Loi de probabilité de la ligne H
df_origine.insert(15,'H_Int_B100',st.gengamma.rvs(1.274647268454995, 0.4915383911925465, -3.4867379104642266e-29, 0.643369376546743, size=731))

#Loi de probabilité de la ligne P
df_origine.insert(17,'P_Int_B100',st.expon.rvs(-9.273081078094038e-08, 18.625967893302214, size=731))

df_origine.head(10)
```

## Calcul des interruptions pour chaque journée

```python
#si le résultat obtenu pour une journée est inférieur à 1 alors on mets 0 sinon on garde a valeur
df_origine['A_Int_B100']=np.where(df_origine['A_Int_B100']<1,0,df_origine['A_Int_B100'])
df_origine['B_Int_B100']=np.where(df_origine['B_Int_B100']<0,0,df_origine['B_Int_B100'])
df_origine['C_Int_B100']=np.where(df_origine['C_Int_B100']<1,0,df_origine['C_Int_B100'])
df_origine['D_Int_B100']=np.where(df_origine['D_Int_B100']<1,0,df_origine['D_Int_B100'])
df_origine['E_Int_B100']=np.where(df_origine['E_Int_B100']<1,0,df_origine['E_Int_B100'])
df_origine['H_Int_B100']=np.where(df_origine['H_Int_B100']<1,0,df_origine['H_Int_B100'])
df_origine['P_Int_B100']=np.where(df_origine['P_Int_B100']<1,0,df_origine['P_Int_B100'])

#On multipli le résultat obtenu par le nombre de départ d'une journée
df_origine.insert(6,'A_Int',df_origine['A_Int_B100'] * df_origine['Line A'] / 100)
df_origine.insert(9,'B_Int',df_origine['B_Int_B100'] * df_origine['Line B'] / 100)
df_origine.insert(12,'C_Int',df_origine['C_Int_B100'] * df_origine['Line C'] / 100)
df_origine.insert(15,'D_Int',df_origine['D_Int_B100'] * df_origine['Line D'] / 100)
df_origine.insert(18,'E_Int',df_origine['E_Int_B100'] * df_origine['Line E'] / 100)
df_origine.insert(21,'H_Int',df_origine['H_Int_B100'] * df_origine['Line H'] / 100)
df_origine['P_Int']=df_origine['P_Int_B100'] * df_origine['Line P'] / 100

df_origine.head(10)
```

## Somme des interruptions sur les deux ans (2019 - 2020)

```python
#On fait la somme des minutes d'interruptions obtenus
print(df_origine['A_Int'].sum(), df_origine['B_Int'].sum(), df_origine['C_Int'].sum(), df_origine['D_Int'].sum(), df_origine['E_Int'].sum(), df_origine['H_Int'].sum(), df_origine['P_Int'].sum())
```

## Présentation des résultats par mois

```python
df_origine.set_index(df_origine['Date'],inplace=True)

df_month = pd.DataFrame()

df_month['Int_Line_A'] = df_origine['A_Int'].resample('M').sum()
df_month['Int_Line_B'] = df_origine['B_Int'].resample('M').sum()
df_month['Int_Line_C'] = df_origine['C_Int'].resample('M').sum()
df_month['Int_Line_D'] = df_origine['D_Int'].resample('M').sum()
df_month['Int_Line_E'] = df_origine['E_Int'].resample('M').sum()
df_month['Int_Line_H'] = df_origine['H_Int'].resample('M').sum()
df_month['Int_Line_P'] = df_origine['P_Int'].resample('M').sum()

df_month.to_csv('BA_Int_2019-2020_Month.csv', index=True, encoding='utf-8', sep=';', decimal=',')

df_month.head(24)
```

## Présentation des résultats par années

```python
df_year = pd.DataFrame()

df_year['Int_Line_A'] = df_origine['A_Int'].resample('A').sum()
df_year['Int_Line_B'] = df_origine['B_Int'].resample('A').sum()
df_year['Int_Line_C'] = df_origine['C_Int'].resample('A').sum()
df_year['Int_Line_D'] = df_origine['D_Int'].resample('A').sum()
df_year['Int_Line_E'] = df_origine['E_Int'].resample('A').sum()
df_year['Int_Line_H'] = df_origine['H_Int'].resample('A').sum()
df_year['Int_Line_P'] = df_origine['P_Int'].resample('A').sum()

df_year.to_csv('BA_Int_2019-2020_Year.csv', index=True, encoding='utf-8', sep=';', decimal=',')

df_year.head(2)
```
