# Calcul des données nécessaires aux KPI

```python
import pandas as pd
import numpy as np
import os
import datetime
os.chdir("")
```

## Indication du fichier que l'on va utiliser

```python
file_origine='BA_P_donnees_preparees.csv'
```

## Lecture du fichier et intégration dans des Dataframes

```python
#Lecture du fichier origine et mise dans un dataframe
df_origine=pd.read_csv(file_origine,delimiter=';', low_memory=False, encoding='latin-1')
#Affichage des 10 premières lignes du dataframe
df_origine.head(10)
```

## Modification du type des colonnes contenant des dates / heures

```python
#Convertion de la colonne au format date
df_origine['FR1_FECHA'] = pd.to_datetime(df_origine['FR1_FECHA'])

#Convertion de la colonne au format forcée de HH:MM:SS
df_origine['FR1_SALC1'] = pd.to_datetime(df_origine['FR1_SALC1'],format="%H:%M:%S",errors='ignore')
df_origine['FR1_SALC2'] = pd.to_datetime(df_origine['FR1_SALC2'],format="%H:%M:%S",errors='ignore')

#Convertion de la colonne au format forcée de HH:MM:SS
df_origine['SAL_C1'] = pd.to_datetime(df_origine['SAL_C1'],format="%H:%M:%S",errors='ignore')
df_origine['SAL_C2'] = pd.to_datetime(df_origine['SAL_C2'],format="%H:%M:%S",errors='ignore')

#Remplacement des NaN par 00:00:01
df_origine['INT_C1'] = df_origine['INT_C1'].fillna('00:00:01')
df_origine['INT_C2'] = df_origine['INT_C2'].fillna('00:00:01')

#Convertion de la colonne au format string
#df_origine['INT_C1'] = df_origine['INT_C1'].apply(str)
#df_origine['INT_C2'] = df_origine['INT_C2'].apply(str)

#Récupération des 5 premiers caractères
#df_origine['INT_C1'] = np.where(df_origine['PERIODO']!=2017, df_origine['INT_C1'].apply(lambda x: x[0:5]).apply(str),df_origine['INT_C1'].apply(str))
#df_origine['INT_C2'] = np.where(df_origine['PERIODO']!=2017, df_origine['INT_C2'].apply(lambda x: x[0:5]).apply(str),df_origine['INT_C2'].apply(str))

#On rajoute 00: à chaqu fois
#df_origine['INT_C1'] = np.where(df_origine['PERIODO']!=2017,"00:" + df_origine['INT_C1'].apply(str),df_origine['INT_C1'].apply(str))
#df_origine['INT_C2'] = np.where(df_origine['PERIODO']!=2017,"00:" + df_origine['INT_C2'].apply(str),df_origine['INT_C2'].apply(str))

#On transforme au format interval de temps
df_origine['INT_C1'] = pd.to_timedelta(df_origine['INT_C1'])
df_origine['INT_C2'] = pd.to_timedelta(df_origine['INT_C2'])

df_origine.head(10)
```

## On filtre les lignes où les départs n'ont pas été effectués (C1)

```python
df_origine.index.name='NUM'

#On récupère seulement les lignes où les départs ont été effectués
df_filtered = df_origine.mask(lambda x: x['FR1_VIAC1'] == "N")

#Les lignes qui n'ont pas été récupéré sont en NaN donc on supprime ces lignes
df_filtered.dropna(subset=['FR1_VIAC1'], inplace=True)
df_filtered.head(10)
```

## On calcul l'interval réel (C1)

```python


#Calcul de l'intervalle de la colonne FR1_SALC1
df_filtered['I_REELC1']=df_filtered['FR1_SALC1']-df_filtered['FR1_SALC1'].shift(1)

#Conversion de colonne d'intervalle en timedelta
df_filtered['I_REELC1'] = pd.to_timedelta(df_filtered['I_REELC1'])

#si un timedelta est négatif alors on mets la valeur à 0
df_filtered['I_REELC1']=np.where(df_filtered['I_REELC1'] < datetime.timedelta(0), 0, df_filtered['I_REELC1'])

df_filtered['I_REELC1']=np.where(df_filtered['FR1_REGIST']<df_filtered['FR1_REGIST'].shift(1),0,df_filtered['I_REELC1'])

df_filtered.head(10)
```

## On filtre les lignes où les départs n'ont pas été effectués (C2)

```python
#On récupère seulement les lignes où les départs ont été effectués
df_filtered_2 = df_origine.mask(lambda x: x['FR1_VIAC2'] == "N")

#Les lignes qui n'ont pas été récupéré sont en NaN donc on supprime ces lignes
df_filtered_2.dropna(subset=['FR1_VIAC2'], inplace=True)
df_filtered_2.head(10)
```

## On calcul l'interval réel (C2)

```python
#Calcul de l'intervalle de la colonne FR1_SALC2
df_filtered_2['I_REELC2']=df_filtered_2['FR1_SALC2']-df_filtered_2['FR1_SALC2'].shift(1)

#Conversion de colonne d'intervalle en timedelta
df_filtered_2['I_REELC2'] = pd.to_timedelta(df_filtered_2['I_REELC2'])

#si un timedelta est négatif alors on mets la valeur à 0
df_filtered_2['I_REELC2']=np.where(df_filtered_2['I_REELC2'] < datetime.timedelta(0), 0, df_filtered_2['I_REELC2'])

df_filtered_2['I_REELC2']=np.where(df_filtered_2['FR1_REGIST']<df_filtered_2['FR1_REGIST'].shift(1),0,df_filtered_2['I_REELC2'])

df_filtered_2.head(10)
```

## On rassemble les intervalles réels calculés en face des bonnes lignes

```python
df_test =df_filtered['I_REELC1']

df_final = pd.concat([df_origine, df_test], axis=1, ignore_index=False)

df_test =df_filtered_2['I_REELC2']

df_final = pd.concat([df_final, df_test], axis=1, ignore_index=False)

df_final.head(10)
df_final.dtypes
```

## Calcul des KPI C1

```python
df_final['Interrup_C1'] = np.where(np.logical_and(df_final['I_REELC1']>(df_final['INT_C1']+df_final['INT_C1'].shift(1)+datetime.timedelta(minutes=2)),df_final['INT_C1']!=datetime.timedelta(seconds=1)),df_final['I_REELC1']-df_final['INT_C1']-df_final['INT_C1'].shift(1)-datetime.timedelta(minutes=2),0)

df_final['Interval_Exc_C1'] = np.where(np.logical_and(df_final['INT_C1']!=datetime.timedelta(seconds=1), df_final['I_REELC1']>df_final['INT_C1']*1.5),1,0)

df_final['Desp_HP_C1_AM'] = np.where(np.logical_and(df_final['TIPO_DIA']=="HABIL",np.logical_and(df_final['FR1_SALC1']>=datetime.datetime(1900,1,1,8,0,0),df_final['FR1_SALC1']<datetime.datetime(1900,1,1,10,0,0))),1,0)
                                  
df_final['Desp_HP_C1_PM'] = np.where(np.logical_and(df_final['TIPO_DIA']=="HABIL",np.logical_and(df_final['FR1_SALC1']>=datetime.datetime(1900,1,1,18,0,0),df_final['FR1_SALC1']<datetime.datetime(1900,1,1,20,0,0))),1,0)

df_final['Desp_HV_C1'] = np.where(np.logical_and(df_final['TIPO_DIA']=="HABIL",np.logical_and(df_final['FR1_SALC1']>=datetime.datetime(1900,1,1,8,0,0),df_final['FR1_SALC1']<datetime.datetime(1900,1,1,10,0,0))),0,
                                  np.where(np.logical_and(df_final['TIPO_DIA']=="HABIL",np.logical_and(df_final['FR1_SALC1']>=datetime.datetime(1900,1,1,18,0,0),df_final['FR1_SALC1']<datetime.datetime(1900,1,1,20,0,0))),0,1))

df_final.head(10)
```

## Calcul des KPI C2

```python
df_final['Interrup_C2'] = np.where(np.logical_and(df_final['I_REELC2']>(df_final['INT_C2']+df_final['INT_C2'].shift(1)+datetime.timedelta(minutes=2)),df_final['INT_C2']!=datetime.timedelta(seconds=1)),df_final['I_REELC2']-df_final['INT_C2']-df_final['INT_C2'].shift(1)-datetime.timedelta(minutes=2),0)

df_final['Interval_Exc_C2'] = np.where(np.logical_and(df_final['INT_C2']!=datetime.timedelta(seconds=1),df_final['I_REELC2']>df_final['INT_C2']*1.5),1,0)

df_final['Desp_HP_C2_AM'] = np.where(np.logical_and(df_final['TIPO_DIA']=="HABIL",np.logical_and(df_final['FR1_SALC2']>=datetime.datetime(1900,1,1,8,0,0),df_final['FR1_SALC2']<datetime.datetime(1900,1,1,10,0,0))),1,0)

df_final['Desp_HP_C2_PM'] = np.where(np.logical_and(df_final['TIPO_DIA']=="HABIL",np.logical_and(df_final['FR1_SALC2']>=datetime.datetime(1900,1,1,18,0,0),df_final['FR1_SALC2']<datetime.datetime(1900,1,1,20,0,0))),1,0)

df_final['Desp_HV_C2'] = np.where(np.logical_and(df_final['TIPO_DIA']=="HABIL",np.logical_and(df_final['FR1_SALC2']>=datetime.datetime(1900,1,1,8,0,0),df_final['FR1_SALC2']<datetime.datetime(1900,1,1,10,0,0))),0,
                                  np.where(np.logical_and(df_final['TIPO_DIA']=="HABIL",np.logical_and(df_final['FR1_SALC2']>=datetime.datetime(1900,1,1,18,0,0),df_final['FR1_SALC2']<datetime.datetime(1900,1,1,20,0,0))),0,1))

df_final.head(10)
```

## Export des données sous CSV

```python
export_tot='BA_' + df_origine.iloc[1,1] + '_DP_TEST.csv'
print(export_tot)
df_final.to_csv(export_tot,sep=';',index=True)
```
